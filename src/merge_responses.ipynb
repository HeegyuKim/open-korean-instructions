```python
import pandas as pd
from datasets import Dataset
```

```python
gpt3 = pd.read_json("../data/openai_gpt-3.5-turb---ultrafeedback", lines=True)
eeve = pd.read_json("../data/yanolja_EEVE-Korean-Instruct-10.8B-v1.0---ultrafeedback", lines=True)
```

```python
gpt3.set_index("__index_level_0__", inplace=True)
eeve.set_index("__index_level_0__", inplace=True)
gpt3
```

```python
cols = ["source", "instruction", "instruction_ko"]

merged = pd.concat([eeve[cols], gpt3.response.rename("gpt3_response"), eeve.response.rename("eeve_response")], axis=1)
merged
```

```python
ds = Dataset.from_pandas(merged[merged.gpt3_response != merged.eeve_response].dropna())
ds = ds.remove_columns(["__index_level_0__"])
ds.push_to_hub("iknow-lab/ultrafeedback_ko_20k")
```
